## Librerías

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Lectura de datos

```python
datos = pd.read_csv("alturas-pesos.csv")
```

```python
datos.head()
```

## Visualizando los datos

```python
sb.scatterplot(x="Altura", y="Peso", data=datos)
```

```python
# Obtenemos las características (X) y la etiqueta (y) de nuestros datos

X = datos["Altura"]
y = datos["Peso"]

X = X.values.reshape(-1,1)
y = y.values.reshape(-1,1)
```

## Modelo

```python
# Creamos una instancia del modelo de regresión lineal

modelo = LinearRegression()
```

## Entrenamiento

```python
modelo.fit(X, y)
```

## Predicción

```python
y_pred = modelo.predict(X)
```

## Gráfico de los datos con la recta de regresión predicha:

```python
plt.scatter(X, y, label='Datos', color = "blue")
plt.plot(X, y_pred, color='purple', label='Recta de regresión')
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.title('Regresión lineal')
plt.legend()
plt.show()
```

## Coeficiente de determinación $R^2$




Es una medida de qué tan bien se ajustan las predicciones del modelo a los datos reales.

**Nota:** Su valor puede variar entre 0 y 1, donde 1 indica un ajuste perfecto.

```python
# Obtenemos el coeficiente de determinación

modelo.score(X, y)
```
